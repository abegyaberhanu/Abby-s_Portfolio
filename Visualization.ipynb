```python
#import needed libraries
import pandas as pd

#initiate and populate the variable with the data from the file with the date as index
data = pd.read_csv(
    '/content/helsinki-vantaa.txt',
    delim_whitespace=True,
    usecols=["DATE",	"USAF",	"TEMP_F",	"TEMP_C",	"REF_TEMP_C",	"DIFF_C"],
    parse_dates=["DATE"],
    index_col="DATE",
)
```

Select data for the 30-year period (January 1988 to December 2018).
•	Store the selection in a new variable selection
•	Enter your code here:


```python
#sort the dataframe by the date so it is in order
data = data.sort_index()

#initiate and populate the variable with the data from the selected date range
selection = data.loc[(data.index >= "19880101") & (data.index <='20181231')]

print(f'Number of rows: {len(selection)}')
```

```python

#import ploting library
import matplotlib.pyplot as plt

#initiate and populate a variable with only the temp within the data range
selection_temps = selection['TEMP_C']

#specify the different aspects of the plot for the figure variable
figure = selection_temps.plot(
    title='Helsinki-Vantaa Airport',
    xlabel='Time',
    ylabel='Temperature (Celsius)',
    figsize=(16, 8),
    marker='o', markersize=4, color='black', linewidth=2

)

#save the plot as a png
plt.savefig("temp_line_plot.png")

```
